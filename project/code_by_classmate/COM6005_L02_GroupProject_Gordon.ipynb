```python
X = df_mAvgFour.iloc[:, 15:].values
y = df_mAvgFour.iloc[:, 0].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sfs = SFS(LinearRegression(),
          k_features="best",
          fixed_features = (0,),
          forward=False,
          scoring='neg_mean_squared_error',
          cv=20)
sfs = sfs.fit(X_scaled, y)

regr = sm.OLS(y,X_scaled).fit()
mse_full = regr.mse_resid

n,p = X_scaled.shape
nsub = len(sfs.subsets_)
bic_values = np.zeros(nsub)
for k in range(nsub):
   mse = -sfs.subsets_[k+1]['avg_score']
   d = len(sfs.subsets_[k+1]['feature_idx'])
   bic_values[k] = mse / mse_full + np.log(n) * d / n

kmin = np.argmin(bic_values)

print(bic_values)

print("Best subset: ",sfs.subsets_[kmin+1]['feature_names'])
print("MSE Score:",-sfs.k_score_)

plt.plot(bic_values, marker='o')
plt.ylabel("BIC")
plt.xlabel("Model Index")
plt.title("BIC for Different Subsets")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import ElasticNetCV
from sklearn.preprocessing import StandardScaler

X = df_mAvgFour.iloc[:, 15:].values
y = df_mAvgFour.iloc[:, 0].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Do elastic net
model = ElasticNetCV(cv=20, l1_ratio=[.1, .5, .7, .9, .95, .99, 1], max_iter=50000) # LASSO + Ridge selection

model.fit(X_scaled, y)
# model.fit(X, y)

best_alpha = model.alpha_
print("Best alpha:", best_alpha)

coef = model.coef_

# Output the non-zero coefficients

non_zero_indices = np.where(coef != 0)[0]
non_zero_coefs = coef[non_zero_indices]

print("Non-zero Coefficients:")
for idx, value in zip(non_zero_indices, non_zero_coefs):
    print(f"Feature {idx}: {value}")
print("Number of non-zero coefficients:", len(non_zero_indices))
```
